## Function which precomputes all conditional probabilities for each feature in the dataset

```python
def precomp(training_data, alpha):
    prec = []
    positive_data = training_data[training_data['target'] == 1]
    negetive_data = training_data[training_data['target'] == 0]
    
    for column in training_data.columns[:-1]:
        prob = {}
        for value in training_data[column].value_counts().keys():
            prob[value] = []
            prob[value].append((len(negetive_data[negetive_data[column] == value]) + alpha)/(len(negetive_data)+14*alpha))
            prob[value].append((len(positive_data[positive_data[column] == value]) + alpha)/(len(positive_data)+14*alpha))
            prob[-1] = [alpha/(len(negetive_data) + 14*alpha), alpha/(len(positive_data) + 14*alpha)]
        prec.append(prob);
        
    return prec
```

## Function which returns the prior probability for a given class

```python
def prior(training_data, target):
    return len(training_data[training_data['target'] == target])/len(training_data)
```

## Function which returns the conditonal probability for a certain feature value given precomputed values, feature, value of feature and the target class

```python
def conditional_prob(prec, feature, value, target):
    if value in prec[feature]:
        prob = prec[feature][value][target]
        
    else:
        prob = 0
        
    if (prob==0):
        prob = prec[feature][-1][target]
    
    return prob
```

## Function which returns the predicted class for a testing example 

```python
def pred(testing_example, prec, pr1, pr2):
    p1 = pr1
    for i in range(14):
        p1 *= conditional_prob(prec, i, testing_example[i], 1)
    
    p2 = pr2
    for i in range(14):
        p2 *= conditional_prob(prec, i, testing_example[i], 0)
    
    if (p1 > p2):
        return 1
    else:
        return 0
```

## Function which evaluates the model and returns the confusion matrix

```python
def evaluate(testing_set, prec, pr1, pr2):
    c_mat = np.zeros((2, 2))
    for testing_example in testing_set:
        y = testing_example[14]
        h = pred(testing_example, prec, pr1, pr2)
        if y==1 and h==1:
            c_mat[1][1]+=1
            
        elif y==1 and h==0:
            c_mat[1][0]+=1
            
        elif y==0 and h==0: 
            c_mat[0][0]+=1
            
        else:
            c_mat[0][1]+=1
            
    return c_mat
```

## Functions for Evaluation Metrics

```python
def accuracy(c_mat):
    return (c_mat[0][0]+c_mat[1][1])/(c_mat[0][0] + c_mat[0][1] + c_mat[1][0] + c_mat[1][1])*100

def precision(c_mat):
    return (c_mat[1][1])/(c_mat[1][1] + c_mat[0][1])*100

def recall(c_mat):
    return (c_mat[1][1])/(c_mat[1][0] + c_mat[1][1])*100

def f1_score(c_mat):
    prec = precision(c_mat)
    rec = recall(c_mat)
    
    return (2*rec*prec)/(prec + rec)
```

## Naive Bayes Classifier using Laplace Smoothing Technique

```python
accuracy_list = np.zeros(10)
precision_list = np.zeros(10)
recall_list = np.zeros(10)
f1_list = np.zeros(10)
for i in range(10):
    
    training_data = dataset.sample(frac=2/3)
    testing_data = dataset.drop(training_data.index)
    testing_set = testing_data.to_numpy()
    
    prec = precomp(training_data, 100)
    pr1 = prior(training_data, 1)
    pr2 = prior(training_data, 0)
    c_mat = evaluate(testing_set, prec, pr1, pr2)
    
    print(f"Model No.{i+1}")
    print(f"Confusion Matrix:\n{c_mat}")
    print(f"Accuracy = {accuracy(c_mat)}")
    print(f"Precision = {precision(c_mat)}")
    print(f"Recall = {recall(c_mat)}")
    print(f"F1-Score = {f1_score(c_mat)}")
    
    accuracy_list[i] =  accuracy(c_mat)
    precision_list[i] = precision(c_mat)
    recall_list[i] =  recall(c_mat)
    f1_list[i] = f1_score(c_mat)
    print("\n")
    
print(f"Accuracy:\n Mean = {accuracy_list.mean()}  Standard Deviation = {accuracy_list.std()}")
print(f"Precision:\n Mean = {precision_list.mean()}  Standard Deviation = {precision_list.std()}")
print(f"Recall:\n Mean = {recall_list.mean()}  Standard Deviation = {recall_list.std()}")
print(f"F1-Score:\n Mean = {f1_list.mean()}  Standard Deviation = {f1_list.std()}")
```
